```python

import random

# 요일별 리스트 생성
#조건에 맞게 선별된 숫자들이 이 요일별 리스트에 저장된다


# 숫자와 문자 대응 딕셔너리 
#조건에 맞게 요일별 리스트에 숫자들이 할당되면 그 숫자에 해당되는 과목이 마지막에 출력된다
number_to_letter = {
    1: "확통",
    2: "딥러닝",
    3: "ai알고",
    4: "기계학습",
    5: "ai수학",
    6: "논문1",
    7: "ai기초",
    8: "서비스",
    9: "논문2",
    10: "영상이해",
    11: "ai코딩",
    12: "리눅스",

}
def trash_cleaner():#함수추가 했지여
    
    mon = []
    tue = []
    wed = []
    thu = []
    fri = []

    # 1부터 12까지의 숫자를 랜덤으로 선택하여 요일별 리스트에 추가 (연속된 요일 간에 겹치지 않게)
    while len(mon) < 5: #5개가 리스트에 추가될때 까지 반복
        random_number = random.randint(1, 12) #1~12 까지 숫자를 랜덤 배정
        if random_number not in mon:   #월요일 리스트는 비었으므로
            mon.append(random_number)  #5번 뽑은 랜덤 숫자가 이미 mon 리스트에 있는지 확인한 후, 만약 없다면 mon 리스트에 추가

    while len(tue) < 5:
        random_number = random.randint(1, 12)
        if random_number not in mon and random_number not in tue:  #월요일괴 화요일에 뽑은 숫자가 없으면 화요일에 숫자를 추가
            tue.append(random_number) 


    while len(wed) < 5:  #수요일 5교시는 진로가 고정이므로 4개만 뽑는다
        random_number = random.randint(1, 12)
        if random_number not in tue and random_number not in wed:
            wed.append(random_number)

    while len(thu) < 5: #이제부턴 반복
        random_number = random.randint(1, 12)
        if random_number not in wed and random_number not in thu:
            thu.append(random_number)

    while len(fri) < 5: #반복
        random_number = random.randint(1, 12)
        if random_number not in thu and random_number not in fri:
            fri.append(random_number)
    #번호 할당 완료
    
    return [mon, tue, wed, thu, fri]




# if trash in mon or trash in tue or trash in thu trash in fri:
while True:
    
    new = trash_cleaner()

    trash=new[2].pop()
    if trash in new[0] or trash in new[1] or trash in new[3] or trash in new[4]:
        trash_cleaner()
    else:
        break




print(new[0])#월요일 일련번호
print(new[1])#화요일 일련번호
print(new[2])#수요일 일련번호
print(new[3])#목요일 일련번호
print(new[4])#금요일 일련번호
print('')

print(trash)


    
    


# 결과 출력
#위에서 할당한 숫자들을 문자 대응 딕셔너리에 맞게 출력한다
print('         ','1교시','  ', ' 2교시','  ', ' 3교시','  ', ' 4교시','  ', ' 5교시')
print("월요일:", [number_to_letter[num] for num in new[0]])
print("화요일:", [number_to_letter[num] for num in new[1]])
print("수요일:", [number_to_letter[num] for num in new[2]] + ["진로"])# 수요일 5교시는 진로
print("목요일:", [number_to_letter[num] for num in new[3]])
print("금요일:", [number_to_letter[num] for num in new[4]])
print('')

```
